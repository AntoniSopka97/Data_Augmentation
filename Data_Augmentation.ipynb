```python
def plot_images(images_arr, batch_size = 10):
    fig, axes = plt.subplots(1,batch_size, figsize = (20,20))
    axes = axes.flatten()
    for img , ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
        
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range = 10,
                        width_shift_range = 0.1,
                        height_shift_range = 0.1,
                        shear_range = 0.15,
                        zoom_range = 0.1,
                        channel_shift_range = 10,
                        horizontal_flip = True)
```

```python
images_name = random.choice(os.listdir('Dataset/dog'))
```

```python
img_path = 'Dataset/dog/' + images_name
```

```python
image = np.expand_dims(plt.imread(img_path),0)
plt.imshow(image[0])
```

```python
# Generate batches of aug images from this image
aug_iter = gen.flow(image)
```

```python
# Get 10 samples
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plot_images(aug_images)
```
